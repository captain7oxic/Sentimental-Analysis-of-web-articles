```python
import nltk
import nltk.data
nltk.download('punkt')
```

```python
#average sentence length calculation

average_sentence_length= []
for essay in input_url_df['Text']:
    tokenized_sentences =  nltk.sent_tokenize(essay)
    tokenized_words = nltk.word_tokenize(essay)
    average_sentence_length.append(len(tokenized_words)/len(tokenized_sentences))
```

```python
#complex Words calculations by calculating syllables

def syllables(word):
    negatives = ["doesn't", "isn't", "shouldn't", "couldn't","wouldn't"]
    word = word.lower()
    syllable_count = 0
    vowels = 'aeiouy'
    if word[0] in vowels:
        syllable_count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            syllable_count += 1
            
    if word.endswith('e'):
        syllable_count -= 1
        
    if word.endswith('le') and len(word) > 2 and word[-3] not in vowels:
        syllable_count += 1
        
    if word[:3] == 'tri' and word[3] in vowels:
        syllable_count +=1
        
    if word[:2] == 'bi' and word[2] in vowels:
        syllable_count +=1
        
    if word[-3:]=="n't":
        if word in negatives:
            syllable_count+=1
        else :
            pass
        
    if word[:2]=='mc':
        syllable_count+=1

    if syllable_count == 0:
        syllable_count += 1
    return syllable_count
```

```python
cw_list =[]
for essay in input_url_df['Text']:
    tokenized_words = nltk.word_tokenize(essay)
    complex_word_count = 0
    for word in tokenized_words:
        if (syllables(word)>2):
            complex_word_count+=1
    cw_list.append(complex_word_count)
        
```

```python
#average Words per sentence
total_sentences = []
for essay in input_url_df["Text"]:
    tokenized_sentences = nltk.sent_tokenize(essay)
    total_sentences.append(len(tokenized_sentences))

avg_words_per_sentence = [x/y for x,y in zip(total_words,total_sentences)]
```

```python
# calculate syllable count per word
# as unaware of the formula i am doing (total number of syllable in Text / total number of words)
total_syllable_count=[]
for essay in input_url_df["Text"]:
    tokenized_words = nltk.word_tokenize(essay)
    syl_count=0
    for word in tokenized_words:
        syl_count+=syllables(word)
    
    total_syllable_count.append(syl_count)
```

```python
# calculations of pronouns
#\b is the boundary
#?-i:us makes the us as case sensitive , while all others remain case insensitive
import re
pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
total_pronoun_count=[]
for essay in input_url_df["Text"]:
    pronouns= pronounRegex.findall(essay)
    total_pronoun_count.append(len(pronouns))   
```

```python
import string
punctuations = set(string.punctuation)
cleaned_words = []
for essay in input_url_df["Text"]:
    filtered = ""
    tokenized_words = nltk.word_tokenize(essay)
    for token in tokenized_words:
        if token not in stopwords_set and token not in punctuations:
            filtered+=token+" "
    cleaned_words.append(filtered)
```
